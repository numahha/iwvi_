```python
import numpy as np
import torch
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D


#https://chrisorm.github.io/VI-MC-PYT.html
```

```python

class VI(torch.nn.Module):
    def __init__(self, x_dim, z_dim, y_dim):
        super(VI, self).__init__()
        self.nu = 1e-6
        self.q_mu = torch.nn.Parameter(torch.zeros(z_dim), requires_grad=True)
        self.q_logvar = torch.nn.Parameter(torch.zeros(z_dim), requires_grad=True) # 対角分散行列
        self.prior_mu = torch.nn.Parameter(torch.zeros(z_dim), requires_grad=False)
        self.prior_logvar = torch.nn.Parameter(torch.zeros(z_dim), requires_grad=True) # 対角分散行列
        self.likelihood_logvar = torch.nn.Parameter(torch.zeros(y_dim), requires_grad=True) # 対角分散行列
        #self.lvr = LatentVariableRegressionModel(x_dim+z_dim, y_dim)
        self.lvr = torch.nn.Sequential(
                                        torch.nn.Linear(x_dim+z_dim, 64),
                                        torch.nn.ReLU(),
                                        torch.nn.Linear(64, 64),
                                        torch.nn.ReLU(),
                                        torch.nn.Linear(64, 64),
                                        torch.nn.ReLU(),
                                        torch.nn.Linear(64, y_dim),
                                    )

    def sample_z(self):
        # reparametrization trick
        std = torch.exp(0.5 * self.q_logvar)
        eps = torch.randn_like(self.q_logvar)
        return eps.mul(std).add_(self.q_mu)        
    
    def gaussian_likelihood_loss(self, y, mu, logvar):
        return 0.5 * torch.mean(((y-mu)**2) * torch.exp(-logvar) + logvar)
        # 分散行列Varが対角成分var_iの対角行列の場合には、log(det|Var|) = log(prod_i var_i) = sum_i log(var_i) 
    
    def compute_loss(self, x, y):
        z = self.sample_z() * torch.ones(x.shape[0],self.q_mu.shape[0])
        xz = torch.cat((x,z),dim=-1)
        y_pred = self.lvr(xz)
        rec_loss = self.gaussian_likelihood_loss(y, y_pred, self.likelihood_logvar)
        kld_loss = self.gaussian_likelihood_loss(z, self.q_mu, self.q_logvar) - self.gaussian_likelihood_loss(z, self.prior_mu, self.prior_logvar)
        #return self.nu*kld_loss - rec_loss
        return rec_loss - self.nu*kld_loss
```

```python

def fn(xz):
    xz = xz.flatten()
    ret = 0
    #ret += xz[0] + xz[1]
    ret += xz[0] * xz[0] *xz[1] 
    return ret + 0*np.random.randn()

def fns(xz):
    xz = xz.reshape(-1,2)
    ret = np.zeros(len(xz))
    for i in range(len(xz)):
        ret[i] = fn(xz[i])
    return ret

x =2*np.random.randn(1000,1)
z = np.linspace(-4,4,20)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

dataset_xz = []
dataset_y = []
for i in range(len(z)):
    tmp_xz = np.hstack([x,z[i]*np.ones(x.shape)])
    #print(tmp_xz)
    tmp_y = -fns(tmp_xz)
    dataset_xz.append(tmp_xz)
    dataset_y.append(tmp_y)
    ax.scatter(tmp_xz[:,0], tmp_xz[:,1], tmp_y)
```

```python
vi = VI(x_dim=1,z_dim=1,y_dim=1)

optimizer = torch.optim.Adam(vi.parameters(),lr=0.1)

for i in range(5000):
    j = np.random.randint(len(dataset_xz))
    x = dataset_xz[j][:,:1]
    y = dataset_y[j]
    x = torch.Tensor(x)[idx[:100]]
    y = torch.Tensor(y)[idx[:100]]
    
    loss = vi.compute_loss(x, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print(i,loss.item())
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

for i in range(len(dataset_xz)):
    xz = dataset_xz[i]
    xz = torch.Tensor(xz)
    xz[:,1] = xz[:,1]*8
    with torch.no_grad():
        y_pred = vi.lvr(xz).numpy()
        
    ax.scatter(xz[:,0], xz[:,1], y_pred) 
```

```python
#reg = LatentVariableRegressionModel(2,1)
reg = torch.nn.Sequential(
        torch.nn.Linear(2, 64),
        torch.nn.ReLU(),
        torch.nn.Linear(64, 64),
        torch.nn.ReLU(),
        torch.nn.Linear(64, 64),
        torch.nn.ReLU(),
    torch.nn.Linear(64, 1),
    )
optimizer = torch.optim.Adam(reg.parameters(),lr=0.01)
loss_func = torch.nn.MSELoss()
for i in range(2000):
    j = np.random.randint(len(dataset_xz))
    # = i % len(dataset_xz)
    xz = dataset_xz[j]
    y = dataset_y[j]
    idx = np.arange(0, len(y), 1)
    np.random.shuffle(idx)
    xz = torch.Tensor(xz) [idx[:100]]
    y = torch.Tensor(y)[idx[:100]]
    
    y_pred = reg(xz)
    loss = loss_func(y.reshape(-1,1), y_pred.reshape(-1,1))#torch.mean((y-y_pred)**2)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print(i,loss.item())
```
