```python

def fn(xz):
    xz = xz.flatten()
    ret = 0
    #ret += xz[0] + xz[1]
    #ret += xz[0] *xz[1]
    ret += np.sin(xz[0] +0*xz[1])
    return ret + 0*np.random.randn()

def fns(xz):
    xz = xz.reshape(-1,2)
    ret = np.zeros(len(xz))
    for i in range(len(xz)):
        ret[i] = fn(xz[i])
    return ret

x =2*np.random.randn(1000,1)
z = np.linspace(-4,4,20)

fig = plt.figure(figsize=(10,10))
#ax = fig.add_subplot(111, projection='3d')
ax = fig.add_subplot(111)

dataset_xz = []
dataset_y = []
for i in range(len(z)):
    tmp_xz = np.hstack([x,z[i]*np.ones(x.shape)])
    tmp_y = -fns(tmp_xz)
    dataset_xz.append(tmp_xz)
    dataset_y.append(tmp_y)
    #ax.scatter(tmp_xz[:,0], tmp_xz[:,1], tmp_y)
    ax.plot(tmp_xz[:,0], tmp_y,"o")
```

```python

class VI(torch.nn.Module):
    def __init__(self, x_dim, z_dim, y_dim):
        super(VI, self).__init__()
        self.nu = 1e-2
        self.x_dim = x_dim
        self.z_dim = z_dim
        self.y_dim = y_dim
        self.q_mu = torch.nn.Parameter(torch.zeros(z_dim))
        self.q_logvar = torch.nn.Parameter(torch.zeros(z_dim))
        self.q_mu_test = torch.nn.Parameter(torch.zeros(z_dim))
        self.q_logvar_test = torch.nn.Parameter(torch.zeros(z_dim))
        self.prior_mu = torch.nn.Parameter(torch.zeros(z_dim))
        self.prior_logvar = torch.nn.Parameter(torch.zeros(z_dim))
        self.likelihood_logvar = torch.nn.Parameter(torch.zeros(y_dim))
        self.lvr = torch.nn.Sequential(
                                        torch.nn.Linear(x_dim+z_dim, 64),
                                        torch.nn.ReLU(),
                                        torch.nn.Linear(64, 64),
                                        torch.nn.ReLU(),
                                        torch.nn.Linear(64, 64),
                                        torch.nn.ReLU(),
                                        torch.nn.Linear(64, y_dim),
                                    )
        self.train_mode()
        
    def train_mode(self):
        self.q_mu.requires_grad=True
        self.q_logvar.requires_grad=True
        self.prior_mu.requires_grad=False
        self.prior_logvar.requires_grad=True
        self.likelihood_logvar.requires_grad=True
        for param in self.lvr.parameters():
            param.requires_grad = True
        self.q_mu_test.requires_grad=False
        self.q_logvar_test.requires_grad=False

    def test_mode(self):
        self.q_mu.requires_grad=False
        self.q_logvar.requires_grad=False
        self.prior_mu.requires_grad=False
        self.prior_logvar.requires_grad=False
        self.likelihood_logvar.requires_grad=False
        for param in self.lvr.parameters():
            param.requires_grad = False
        self.q_mu_test.requires_grad=True
        self.q_logvar_test.requires_grad=True            
    
    def gaussian_likelihood_loss(self, y, mu, logvar):
        return 0.5 * torch.mean(((y-mu)**2) * torch.exp(-logvar) + logvar)
        # 分散行列Varが対角成分var_iの対角行列の場合には、log(det|Var|) = log(prod_i var_i) = sum_i log(var_i) 
    
    def kld(self, mu1, logvar1, mu2, logvar2):
        # kld(p1|p2) = E_{z~p1}[ log p1(z) - log p2(z) ]
        tmp1 = 0.5 * (logvar2 - logvar1) # log (sigma2/sigma1)
        tmp2 = 0.5*(torch.exp(logvar1)+(mu1-mu2)**2) / torch.exp(logvar2) # (sigma1^2+(mu1-mu2)^2)/(2*sigma2^2)
        return torch.mean(tmp1 + tmp2)
    
    def compute_loss(self, x, y):

        x = x.reshape(-1,self.x_dim)
        y = y.reshape(-1,self.y_dim)

        # reparametrization trick
        eps = torch.randn_like(self.q_logvar)
        std = torch.exp(0.5 * self.q_logvar)
        z = (eps*std+self.q_mu) * torch.ones(x.shape[0],self.z_dim)

        xz = torch.cat((x,z),dim=-1)
        
        y_pred = self.lvr(xz)

        loss = 0
        loss += self.gaussian_likelihood_loss(y, y_pred, self.likelihood_logvar) # approx of E_{z~q}[ - log p(y|x,z) ]
        loss += self.nu * self.kld(self.q_mu, self.q_logvar, self.prior_mu, self.prior_logvar) # nu * E_{z~q}[ log q(z) - log p(z) ]
        return loss

#         loss += self.nu * self.gaussian_likelihood_loss(z, self.prior_mu, self.prior_logvar) # nu *[- log p(z)]
#         if train_flag:
#             loss -= self.nu * self.gaussian_likelihood_loss(z, self.q_mu, self.q_logvar) # nu * [- log q(z)]
#         else:
#             loss -= self.nu * self.gaussian_likelihood_loss(z, self.q_mu_test, self.q_logvar_test) # nu * [- log q(z)]
        #return loss


    
vi = VI(x_dim=1,z_dim=1,y_dim=1)
#vi.test_mode()
#print(list(vi.parameters()))

optimizer = torch.optim.Adam(vi.parameters(),lr=0.001)


for i in range(2000):
    j = np.random.randint(len(dataset_xz))
    x = dataset_xz[j][:,:1]
    y = dataset_y[j]
    x = torch.Tensor(x)#[idx[:100]]
    y = torch.Tensor(y)#[idx[:100]]
    optimizer.zero_grad()
    loss = vi.compute_loss(x, y)
    loss.backward()
    optimizer.step()
    print(i,loss.item())
```
